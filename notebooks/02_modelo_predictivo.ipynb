```python
import pandas as pd
from scipy.stats import mannwhitneyu

df = pd.read_csv('../data/nyc_taxi_weather.csv')

# Comparar duración en días con lluvia y los días despejados
rain = df[df['clima_main'] == 'Rain']['duration_min']
clear = df[df['clima_main'] == 'Clear']['duration_min']

stat, p = mannwhitneyu(rain, clear)
print("Estadístico:", stat)
print("p-valor:", p)

if p < 0.05:
    print("Hay diferencia significativa entre días de lluvia y días despejados.")
else:
    print("No hay diferencia significativa.")
```
